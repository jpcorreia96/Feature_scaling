## Trazendo as features para a mesma escala

 Trazer as features para a mesma escala é uma etapa crucial no pré-processamento de dados em projetos de machine learning. A normalização e a padronização são dois métodos comumente usados para esse propósito. Vamos explorar a importância dessa prática e as duas técnicas principais.

### Importância de Escalar as Features

*   Muitos algoritmos de machine learning, como a regressão linear, redes neurais e SVMs, usam métodos de otimização que convergem mais rapidamente quando as features estão na mesma escala.

*   Outros algoritmos que dependem de cálculos de distância, como K-Nearest Neighbors (KNN) e clustering (ex.: K-Means), são particularmente sensíveis à escala das features. Features em escalas diferentes podem dominar a medida de distância, prejudicando a performance do modelo.


### Normalização
*   A normalização transforma as features para um intervalo específico, geralmente [0, 1].

$$ x_{norm}^{(i)} = \dfrac{x^{(i)} - x_{min}}{x_{max} - x_{min}}$$

onde $x^{(i)}$ é um exemplo particular, $x_{min}$ é o menor valor da feature na coluna e $x_{max}$ é o maior valor.
#### A normalização geralmente é  

*  Utilizada quando os dados não têm uma distribuição gaussiana (normal).

*  Mantém a relação de dispersão dos dados.

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X_train_norm = mms.fit_transform(X_train)
X_test_norm = mms.fit_transform(X_test)
```

```python
def plot_histograms(data, columns, title_prefix, num_cols=3):
    num_rows = (len(columns) + num_cols - 1) // num_cols
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))
    axes = axes.flatten()

    # Plotar histogramas para cada coluna de dados normalizados
    for i, col in enumerate(columns):
        sns.histplot(data[:, i], bins=10, kde=True, ax=axes[i])
        axes[i].set_title(f'{title_prefix} {col}')

    # Remover eixos vazios
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()

# Plotar histogramas dos dados normalizados (conjunto de treino)
plot_histograms(X_train_norm, X_train.columns, 'Histograma Normalizado da')

```

### Padronização

A padronização transforma as features para terem média 0 e desvio padrão 1.

$$x_{std}^{(i)} =\dfrac{x^{(i)} - \mu_x}{\sigma_x}$$

onde $\mu_x$ é a média da coluna da feature e $\sigma_x$ é desvio padrão.

####  A padronização geralmente é 

* Utilizada quando os dados têm uma distribuição gaussiana.

* Ajudar a modelar dados que podem ter outliers ou variâncias diferentes.



```python
from sklearn.preprocessing import StandardScaler 
stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.fit_transform(X_test)
```

```python
# Plotar histogramas dos dados normalizados (conjunto de teste)
plot_histograms(X_train_norm, X_train.columns, 'Histograma Normalizado da')
```
